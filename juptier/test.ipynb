```python
import numpy as np
import matplotlib.pyplot as plt

# Load a few chips to verify quality
t0_chip = np.load("C:/Users/Rahul/CollegeProject/UrbanEyeML/urbeye-ml/data/chips/s2_t0_0_0.npy")  # Shape: (4, 256, 256)
t1_chip = np.load("C:/Users/Rahul/CollegeProject/UrbanEyeML/urbeye-ml/data/chips/s2_t1_0_0.npy")

# Visualize RGB composite
def show_rgb(chip, title):
    rgb = chip[[2,1,0], :, :].transpose(1,2,0)  # Red, Green, Blue
    rgb = np.clip(rgb / 3000, 0, 1)  # Normalize for display
    plt.imshow(rgb)
    plt.title(title)

plt.figure(figsize=(12, 5))
plt.subplot(1,2,1); show_rgb(t0_chip, "Before (t0)")
plt.subplot(1,2,2); show_rgb(t1_chip, "After (t1)")
plt.show()
```
